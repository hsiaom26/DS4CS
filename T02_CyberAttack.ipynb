### How do we identify an attack?
* (Misuse Detection) v.s. (Anomaly Detection)
  * white list, black list
* How to collect evidence? Active, interactive, passive
  * Fingerprinting
  * Network Telescope
  * Honeypot
    * low- and high-interactive
    * file, email, 
  * auditd, syslog, Windows Event
  * packet capture, packet mirror
* Based on what? (Carrier)
  * File: Antivirus
  * Packet: Firewall, (Intrusion Detection System)
  * Application Layer Filtering: HTTP URL, SMTP mailto
* When?
  * (Intrusion Detection System), IDS
  * Intrusion Prevenstion System, IPS
  * Post-Mortem Forensics
  * Lab Reproduce

  * Host-based
